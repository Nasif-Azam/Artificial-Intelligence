```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
os.chdir("/content")
data = pd.read_csv("demo.csv")
```

```python
X = data['X']
Y = data['Y']
Z = data['Z']

plt.plot(X, Y, label='Y', linestyle='-.')
plt.plot(X, Z, label='Z', linestyle='-')

plt.legend()

plt.xlabel("marks")
plt.ylabel("Person")
```
