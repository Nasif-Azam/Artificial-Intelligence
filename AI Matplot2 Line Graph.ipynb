**Necessary Libraries**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

**Real World Examples**

```python
data = pd.read_csv("https://raw.githubusercontent.com/KeithGalli/matplotlib_tutorial/master/gas_prices.csv")
plt.figure(figsize=(16,8)) # Figure method is always mentioned in top or after the CSV file load 
plt.title('Gas Prices in various country', fontdict={'fontname': 'Times New Roman', 'fontsize': '20', 'fontweight': 'bold'})
# print(data)

years = data['Year']
# print(years[::3])
# usa = data['USA']
# canada = data['Canada']
south_korea = data['South Korea']

# Plot graphs values manually
# plt.plot(years, data.USA, 'r.-', linewidth=2.5, markersize=10, label='USA' )
# plt.plot(years, data.Canada, label='Canada', linestyle='-.', color='brown', linewidth=2.5, marker='.', markersize=10, markeredgecolor='brown')
# plt.plot(years, south_korea, label='South Korea', linestyle='-.', color='orange', linewidth=2.5, marker='.', markersize=10, markeredgecolor='black')
# plt.plot(years, data.Australia, 'g.-', linewidth=2.5, markersize=10, label='Australia' )
# plt.plot(years, data.France, 'y.-', linewidth=2.5, markersize=10, label='France')
# plt.plot(years, data.Germany, 'b.-', linewidth=2.5, markersize=10, label='Germany')
# plt.plot(years, data.Italy, 'c.-.', linewidth=2.5, markersize=10, label='Italy')
# plt.plot(years, data.Japan, 'k.--', linewidth=2.5, markersize=10, label='Japan')

# Plot graphs values automatically
for country in data:
  if(country != 'Year'):
    plt.plot(years, data[country], linewidth=2.5, marker='.', markersize=10, label=country)

plt.xticks(years[::3].tolist()+[2011]) # tolist() is used to make data frame to list. As we can extend years to 2011

plt.xlabel('Years')
plt.ylabel('US Dollar')

plt.legend()

# plt.savefig('Gas Price Figure.png', dpi=300) # To Save graph

plt.show()
```
