**Necessary Libraries**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

**Matplotlib Basics**

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]
z = [5,3,7,8,9]
plt.figure(figsize=(10,6)) #Figure method is always mentioned in top or after the CSV file load
plt.title('My Graph', fontdict={'fontname': 'Times New Roman', 'fontsize': '20', 'fontweight': 'bold'})
plt.plot(x,y, 'r.-', label='X', linewidth=2.5, markersize=15 )
plt.plot(x,z, label='Z', linestyle='-', color='gold', linewidth=2.5, marker='^', markersize=10, markeredgecolor='black' )
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([0,2,4,6,6.5])
plt.yticks([0,1,2,3,4,6,7,8,9,10,10.5])

plt.legend()
plt.show()
```
