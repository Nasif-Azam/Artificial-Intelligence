**Necessary Libraries**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/matplotlib_tutorial/master/fifa_data.csv')
# plt.figure(figsize=(16,8)) # Figure method is always mentioned in top or after the CSV file load 
# print(data.head(5))
data['Preferred Foot']

left = data.loc[data['Preferred Foot'] == 'Left'].count()[0]
right = data.loc[data['Preferred Foot'] == 'Right'].count()[0]

plt.title('Distribution of Player Skills in FIFA 2018', fontdict={'fontname': 'Times New Roman', 'fontsize': '20', 'fontweight': 'bold'})
plt.pie([left, right], labels=['Left', 'Right'], colors=['olive', 'skyblue'], autopct='%.3f%%')

plt.show()
data['Weight']
  
```

```python
# plt.figure(figsize=(16,8)) # Figure method is always mentioned in top or after the CSV file load 
plt.style.use('ggplot')

data.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in data.Weight]

light = data.loc[data.Weight < 125].count()[0]
light_medium = data[(data.Weight >= 125) & (data.Weight < 175)].count()[0]
medium = data[(data.Weight >= 150) & (data.Weight < 175)].count()[0]
medium_heavy = data[(data.Weight >= 175) & (data.Weight < 200)].count()[0]
heavy = data[(data.Weight >= 200)].count()[0]

plt.title('Distribution of Player Skills in FIFA 2018', fontdict={'fontname': 'Times New Roman', 'fontsize': '20', 'fontweight': 'bold'})
# colors = ['black', 'blue', 'red', 'green', 'yellow'] # User defined colors
explode = (.4,0,0,0,.4)
plt.pie([light, light_medium, medium, medium_heavy, heavy], labels=['Light', 'Light Medium', 'Medium', 'Medium Heavy', 'Heavy'], autopct='%.3f%%', pctdistance=0.8, explode=explode)

plt.show()
```
